## Lets get started 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sk


# get the data from the csv file
X_train = pd.read_csv('./data/X_train_J01Z4CN.csv')
print(X_train.shape)

prices = pd.read_csv('./data/y_train_OXxrJt1.csv')
```

```python
X_train.head()
```

```python
prices.head()
```

```python
# merge the two dataframes by id_annonce 
df = pd.merge(X_train, prices, on='id_annonce')

# now drop the id_annonce column
df.drop('id_annonce', axis=1, inplace=True)

df.head()
```

```python
# get types of each column
# print(df.dtypes)

# get column names for numerical and categorical columns
num_cols = df._get_numeric_data().columns
cat_cols = list(set(df.columns) - set(num_cols))

print(cat_cols)
print(num_cols)


```

```python
# check for missing values
print((df.isnull().sum())/len(df))
```

Here i'll start first with a model where i drop columns with more that 70% missing values

```python
threshold = 0.7

missing_values_percent = (df.isnull().sum())/len(df)

columns_to_drop = missing_values_percent[missing_values_percent > threshold].index

df.drop(columns_to_drop, axis = 1, inplace = True)
```

## Exploratory Data Analysis

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2, random_state=42)

# this is because train_test_split returns a list and we need a dataframe
train = pd.DataFrame(train)
test = pd.DataFrame(test)

train
```
